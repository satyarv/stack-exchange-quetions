```python
from bs4 import BeautifulSoup
import re

def cleaner(text):
    # 1. Remove HTML tags
    text = BeautifulSoup(text, "html.parser").get_text()

    # 2. Keep only alphabetic characters
    text = re.sub("[^a-zA-Z]", " ", text)

    # 3. Convert to lowercase
    text = text.lower()

    # 4. Remove extra spaces
    text = re.sub(r"\s+", " ", text)

    # 5. Create SpaCy doc object
    doc = nlp(text)

    # 6. Remove stopwords and lemmatize the words
    tokens = [token.lemma_ for token in doc if not token.is_stop]

    # 7. Join tokens back to string
    return " ".join(tokens)
```

```python
freq = {}
for i in df['tags']:
    for j in i:
        if j in freq:
            freq[j] += 1
        else:
            freq[j] = 1
```

```python
common_tags = list(freq.keys())[:10]
common_tags
```

```python
x = []
y = []

for i in range(len(df['tags'])):
    temp = []
    for j in df['tags'][i]:
        if j in common_tags:
            temp.append(j)

    if len(temp) > 1:
        x.append(df['cleaned_text'][i])
        y.append(temp)
```

```python
def classify(pred_prob, threshold):
    y_pred_seq = []

    for i in pred_prob:
        temp = []
        for j in i:
            if j >= threshold:
                temp.append(1)
            else:
                temp.append(0)
        y_pred_seq.append(temp)

    return y_pred_seq
```

```python
import numpy as np
from sklearn.metrics import f1_score

def optimum_threshold(actual, pred_prob):
    # define candidate threshold values from 0.0 to 0.5 with step 0.01
    thresholds = np.arange(0, 0.5, 0.01)
    score = []

    for value in thresholds:
        # Get binary predictions for this threshold
        pred_classes = classify(pred_prob, value)
        # Compute F1-score
        score.append(f1_score(actual, pred_classes, average="weighted"))

    # Return threshold with the highest F1-score
    return thresholds[score.index(max(score))]
```

```python
optimal_threshold = optimum_threshold(y_tr, train_pred_nb)
print("Optimal threshold:", optimal_threshold)
```

```python
# Importing Logistic Regression
from sklearn.linear_model import LogisticRegression

# Defining Model
lr_model = OneVsRestClassifier(LogisticRegression())

# Training Model
lr_model.fit(train_word_features, y_tr)

# Make Predictions on Train Set
train_pred_lr = lr_model.predict_proba(train_word_features)

# Finding Optimum Threshold
print("Optimal threshold =>", optimum_threshold(y_tr, train_pred_lr))

# Getting classes using optimum threshold (Assuming threshold found is 0.34)
train_pred_lr_class = classify(train_pred_lr, 0.34)
print(train_pred_lr_class[:5])

# Evaluating on Training Set
print("F1-score on Train Set:", f1_score(y_tr, train_pred_lr_class, average="weighted"))

# Make Predictions on Validation Set
val_pred_lr = lr_model.predict_proba(test_word_features)

# Getting Classes
val_pred_lr_class = classify(val_pred_lr, 0.34)

# Evaluating on Validation Set
print("F1-score on Validation Set:", f1_score(y_val, val_pred_lr_class, average="weighted"))
```

```python
def tagging(question):
    # Text Cleaning
    cleaned_question = cleaner(question)

    # Feature Engineering
    vector = word_vectorizer.transform([cleaned_question])

    # Predicting Probabilities
    pred_prob = lr_model.predict_proba(vector)

    # Converting Probabilities into Classes
    pred_class = classify(pred_prob, 0.34)

    # Returning the predicted tags
    return mlb.inverse_transform(np.array(pred_class))
```

```python
question = """
I'm using SVM classification (Matlab) within my research works, and I want to know:
1. The advantages and disadvantages of each training algorithm, i.e., SMO, LS and QP
2. In general case, what is the suitable algorithm?
3. If the choice of one of them can impact the classification performance?
4. Is there any relationship between the choice of training method and the kernel function?
"""

predicted_tags = tagging(question)
print("Predicted Tags:", predicted_tags)
```
